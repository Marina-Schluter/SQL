### Ejercicio 5

#### Obtener por cada uno de los repartidores su DNI junto a la cantidad de pedidos entregados.

```python
connection = mysql.connector.connect(host='35.208.82.175', database='dbgsipp7ucsuwy', user='uqsnolewgvge8', password='upgradehub', connection_timeout=180000)
cursor=connection.cursor()
query5 = """ SELECT DNI_R, COUNT(Numero) AS Pedidos_entregados 
                FROM PEDIDO 
                WHERE Hora_rep IS NOT NULL 
                GROUP BY DNI_R;"""
cursor.execute(query5)
cursor.fetchall()


#Seleccionamos las columnas "DNI_R" y "Numero" de la tabla pedidos
#Como queremos mostrar la suma de elementos en la columna "Numero", aplicamos un "COUNT" antes de añadirla y le decimos cómo queremos llamarla usando "AS nuevo_nombre"
#Pedimos que nos devuelva aquellas entradas donde la hora de recepción está completada usando "WHERE" y "IS NOT NULL"
#Ordenamos por la columna "DNI_R" de forma ascendente usando "GROUP BY"
#"GROUP BY" debe ir siempre al final, cuando la info ya está añadida y filtrada
#"GROUP BY" ordena de forma ascendente como forma predeterminada. Para ordenar de forma descentente añadiríamos "DESC"
#"GROUP BY" puede usarse con el nombre o con el número de la columna
```

```python
df5 = pd.read_sql(query5, con=connection)
df5
```

### Ejercicio 9

```python
connection = mysql.connector.connect(host='35.208.82.175', database='dbgsipp7ucsuwy', user='uqsnolewgvge8', password='upgradehub', connection_timeout=180000)
cursor=connection.cursor()
query9 = """ SELECT e.Nombre, e.DNI 
                FROM EMPLEADO e 
                LEFT JOIN PEDIDO p ON e.DNI = p.DNI_EP 
                WHERE p.DNI_EP IS NULL; """
cursor.execute(query9)
cursor.fetchall()


#Seleccionamos las columnas "Nombre" y "DNI" de la tabla "empleado"
#Como vamos a utilizar más de una tabla, para evitar confusiones, ponemos un distintito antes de cada columa para saber de qué tabla viene. A la tabla "Empleado" le asignamos una "e"
#Unimos la tabla "empleado" con la tabla "pedido" usando un "LEFT JOIN", que nos devolverá todas las filas de la tabla izquierda ("empleado" y las filas coincidentes de la tabla derecha ("pedido")
#Si no hubieran coincidencias, a la derecha devolvería un "NULL" donde no la hubiese
#El valor común o condición de unión es el número de DNI y por eso le decimos que nos devuelva en la columna DNI de empleado, la columna DNI_EP de pedidos
#Filtramos la columna DNI_EP de pedidos (la añadida), para obtener solo aquellos que no tienen coincidencia con la tabla de "pedidos" para ver los que no han preparado pedidos. Usamos "IS NULL"
```

```python
df9 = pd.read_sql(query9, con=connection)
df9
```
